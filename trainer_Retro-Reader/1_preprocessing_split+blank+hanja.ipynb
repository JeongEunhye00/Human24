```python
v_newf= OrderedDict()

v_newf['version']="v2.0"
v_newf['data']=[]

v_span_batch = v_span_batch//10
v_how_batch = v_how_batch//10
v_unanswerable_batch = v_unanswerable_batch//10

# test set
span_start= v_span_batch*0
how_start= v_how_batch*0
unanswerable_start= v_unanswerable_batch*0

span_end= span_start+ v_span_batch
how_end= how_start+ v_how_batch
unanswerable_end= unanswerable_start+ v_unanswerable_batch

for iter in range(10):

  print('iter ', iter)
  print('span_start', span_start, 'span_end', span_end, 'batch', span_end-span_start)
  print('how_start', how_start, 'how_end', how_end, 'batch', how_end-how_start)
  print('unanswerable_start', unanswerable_start, 'unanswerable_end', unanswerable_end, \
        'batch', unanswerable_end-unanswerable_start, '\n\n')  
  
  # span data
  for i in range(span_start, span_end):
    title= v_span_data['data'][i]['doc_title']

    for k in range(len(v_span_data['data'][i]['paragraphs'])):
      temp= dict()
      temp_list= []

      for j in range(len(v_span_data['data'][i]['paragraphs'][k]['qas'])):
        question_temp= v_span_data['data'][i]['paragraphs'][k]['qas'][j]['question']

        # 괄호 띄어쓰기
        re_temp= question_temp.replace('(', " (")
        re_temp= re_temp.replace(')', ") ")
        re_temp= re_temp.replace('{', " {")
        re_temp= re_temp.replace('}', "} ")
        re_temp= re_temp.replace('[', " [")
        re_temp= re_temp.replace(']', "] ")

        # 한자 음독 변환
        re_temp_h= hanja.translate(re_temp, 'substitution')

        # 의미없는 띄어쓰기 삭제

        question= re_temp_h
        id= v_span_data['data'][i]['paragraphs'][k]['qas'][j]['question_id']
        answer_text_temp= v_span_data['data'][i]['paragraphs'][k]['qas'][j]['answers']['text']

        # 괄호 띄어쓰기
        re_temp= answer_text_temp.replace('(', " (")
        re_temp= re_temp.replace(')', ") ")
        re_temp= re_temp.replace('{', " {")
        re_temp= re_temp.replace('}', "} ")
        re_temp= re_temp.replace('[', " [")
        re_temp= re_temp.replace(']', "] ")

        # 한자 음독 변환
        re_temp_h= hanja.translate(re_temp, 'substitution')

        # 의미없는 띄어쓰기 삭제

        answer_text= re_temp_h

        answer_start= v_span_data['data'][i]['paragraphs'][k]['qas'][j]['answers']['answer_start'] # null
        # is_impossible= span_data['data'][i]['paragraphs'][k]['qas'][j]['is_impossible']

        temp= {'question':question, \
              'id':id, \
              'answers':[ \
              {'text': answer_text, \
              'answer_start': answer_start}], \
              'is_impossible': False}
        
        temp_list.append(temp)
    
      context_temp= v_span_data['data'][i]['paragraphs'][k]['context']

      # 괄호 띄어쓰기
      re_temp= context_temp.replace('(', " (")
      re_temp= re_temp.replace(')', ") ")
      re_temp= re_temp.replace('{', " {")
      re_temp= re_temp.replace('}', "} ")
      re_temp= re_temp.replace('[', " [")
      re_temp= re_temp.replace(']', "] ")
      
      # 한자 음독 변환
      re_temp_h= hanja.translate(re_temp, 'substitution')

      # 의미없는 띄어쓰기 삭제

      context= re_temp_h

    v_newf['data'].append({'title':title, 'paragraphs': [{'qas': temp_list, \
                                            'context': context} \
                                            ] \
                })
    
  span_start += v_span_batch
  span_end += v_span_batch


  # unanswerable data
  for i in range(unanswerable_start, unanswerable_end):
    title= v_unanswerable_data['data'][i]['doc_title']

    for k in range(len(v_unanswerable_data['data'][i]['paragraphs'])):
      temp= dict()
      temp_list= []

      for j in range(len(v_unanswerable_data['data'][i]['paragraphs'][k]['qas'])):
        plausible_answers= v_unanswerable_data['data'][i]['paragraphs'][k]['qas'][j]['answers']['text']
        answer_start= v_unanswerable_data['data'][i]['paragraphs'][k]['qas'][j]['answers']['answer_start']
        question_temp= v_unanswerable_data['data'][i]['paragraphs'][k]['qas'][j]['question']

        # 괄호 띄어쓰기
        re_temp= question_temp.replace('(', " (")
        re_temp= re_temp.replace(')', ") ")
        re_temp= re_temp.replace('{', " {")
        re_temp= re_temp.replace('}', "} ")
        re_temp= re_temp.replace('[', " [")
        re_temp= re_temp.replace(']', "] ")

        # 한자 음독 변환
        re_temp_h= hanja.translate(re_temp, 'substitution')

        # 의미없는 띄어쓰기 삭제

        question= re_temp_h

        id= v_unanswerable_data['data'][i]['paragraphs'][k]['qas'][j]['question_id']

        temp= {'plausible_answers': [{'text': plausible_answers, 'answer_start': answer_start}], \
              'question':question, \
              'id':id, \
              'answers':[], \
              'is_impossible': True}
        
        temp_list.append(temp)
    
      context_temp= v_unanswerable_data['data'][i]['paragraphs'][k]['context']

      # 괄호 띄어쓰기
      re_temp= context_temp.replace('(', " (")
      re_temp= re_temp.replace(')', ") ")
      re_temp= re_temp.replace('{', " {")
      re_temp= re_temp.replace('}', "} ")
      re_temp= re_temp.replace('[', " [")
      re_temp= re_temp.replace(']', "] ")
      
      # 한자 음독 변환
      re_temp_h= hanja.translate(re_temp, 'substitution')

      # 의미없는 띄어쓰기 삭제

      context= re_temp_h


    v_newf['data'].append({'title':title, 'paragraphs': [{'qas': temp_list, \
                                            'context': context} \
                                            ] \
                })
    
  unanswerable_start += v_unanswerable_batch
  unanswerable_end += v_unanswerable_batch


  # how data
  for i in range(how_start, how_end):
    title= v_how_data['data'][i]['doc_title']

    for k in range(len(v_how_data['data'][i]['paragraphs'])):
      temp= dict()
      temp_list= []

      for j in range(len(v_how_data['data'][i]['paragraphs'][k]['qas'])):
        question_temp= v_how_data['data'][i]['paragraphs'][k]['qas'][j]['question']

        # 괄호 띄어쓰기
        re_temp= question_temp.replace('(', " (")
        re_temp= re_temp.replace(')', ") ")
        re_temp= re_temp.replace('{', " {")
        re_temp= re_temp.replace('}', "} ")
        re_temp= re_temp.replace('[', " [")
        re_temp= re_temp.replace(']', "] ")

        # 한자 음독 변환
        re_temp_h= hanja.translate(re_temp, 'substitution')

        # 의미없는 띄어쓰기 삭제

        question= re_temp_h

        id= v_how_data['data'][i]['paragraphs'][k]['qas'][j]['question_id']
        answer_text_temp= v_how_data['data'][i]['paragraphs'][k]['qas'][j]['answers']['text']

        # 괄호 띄어쓰기
        re_temp= answer_text_temp.replace('(', " (")
        re_temp= re_temp.replace(')', ") ")
        re_temp= re_temp.replace('{', " {")
        re_temp= re_temp.replace('}', "} ")
        re_temp= re_temp.replace('[', " [")
        re_temp= re_temp.replace(']', "] ")

        # 한자 음독 변환
        re_temp_h= hanja.translate(re_temp, 'substitution')

        # 의미없는 띄어쓰기 삭제

        answer_text= re_temp_h

        answer_start= v_how_data['data'][i]['paragraphs'][k]['qas'][j]['answers']['answer_start'] # null
        # is_impossible= how_data['data'][i]['paragraphs'][k]['qas'][j]['is_impossible']

        temp= {'question':question, \
              'id':id, \
              'answers':[ \
              {'text': answer_text, \
              'answer_start': answer_start}], \
              'is_impossible': False}
        
        temp_list.append(temp)
    
      context_temp= v_how_data['data'][i]['paragraphs'][k]['context']

      # 괄호 띄어쓰기
      re_temp= context_temp.replace('(', " (")
      re_temp= re_temp.replace(')', ") ")
      re_temp= re_temp.replace('{', " {")
      re_temp= re_temp.replace('}', "} ")
      re_temp= re_temp.replace('[', " [")
      re_temp= re_temp.replace(']', "] ")
      
      # 한자 음독 변환
      re_temp_h= hanja.translate(re_temp, 'substitution')

      # 의미없는 띄어쓰기 삭제

      context= re_temp_h

    v_newf['data'].append({'title':title, 'paragraphs': [{'qas': temp_list, \
                                            'context': context} \
                                            ] \
                })

  how_start += v_how_batch
  how_end += v_how_batch
```
